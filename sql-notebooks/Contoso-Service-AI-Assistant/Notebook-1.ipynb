```sql
declare @e vector(1536) 
exec dbo.get_embeddings @text = 'The customer was displeased with timely delivery of the vehicle', @embedding = @e output;


	select 
    sf.feedback_text,   sf.rating_overall_experience,
    vector_distance('cosine', @e, sf.feedback_vector) as distance
from
    Service_Feedback sf
where
    vector_distance('cosine', @e, sf.feedback_vector) < 0.5
	and sf.rating_overall_experience <=3
order by
    distance
```
